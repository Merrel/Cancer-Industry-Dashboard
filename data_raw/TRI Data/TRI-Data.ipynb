```python
# Import package
import pandas as pd
import json
```

```python
a = pd.read_csv("TRI_2016_flowbyfacility.csv")
b = pd.read_csv("TRI_2016_facility.csv")
merged = b.merge(a, on='FacilityID')
```

```python
#merged_GA = merged.loc[merged['State'] == 'GA']
#merged_GA.to_csv("output_GA.csv", index=False)
#merged_GA.head(20)

merged.to_csv("data_US.csv", index=False)
merged.head(20)
```

```python
df = merged.groupby(['County','FlowName'],as_index = False)['FlowAmount'].sum().pivot('County','FlowName').fillna(0)
df.head(20)
```

```python
#df = merged.groupby(['County','FlowName'],as_index = False)['FlowAmount'].sum().pivot('County','FlowName').fillna(0)
df = merged.groupby(['State','County','FlowName'],as_index = False)['FlowAmount'].sum()
df['Region'] = df.County.astype(str).str.cat(df.State.astype(str), sep=', ')
df.head(20)
```

```python
df1 = df.drop(columns=['State', 'County'])
#df1 = df.pivot('Region','FlowName').fillna(0)
df1
```

```python
df1 = df1.pivot('Region','FlowName').fillna(0)
df1
```

```python
df1.columns = df1.columns.droplevel(0) 
df1.columns.name = None              
df1 = df1.reset_index()  
df1
```

```python
df1.to_csv("output_US.csv", index=False)
```
