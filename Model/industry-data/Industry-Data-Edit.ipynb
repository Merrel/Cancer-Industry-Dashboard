```python
# Importing package
import re
import pandas as pd
import numpy as np
```

# Load data

```python
df = pd.read_csv("indicators_per-industry_per-county.csv")
df
```

# Data Processing

```python
# Code is the first three digits of the industry code

df["code"] = df.industry_code.str.slice(0, 3)
df.head(20)
```

```python
df.columns
```

```python
# List of 3-digit industry codes (70 codes)

codes = df.code.unique()
codes
```

```python
df_new = df.groupby(['fips', 'name', 'year', 'code'], as_index=False)['payann', 'estab', 'emp',
    'ACID', 'ENRG', 'ETOX', 'EUTR', 'FOOD', 'GCC', 'HAPS', 'HAZW', 'HC', 
    'HNC', 'HRSP', 'HTOX', 'JOBS', 'LAND', 'METL', 'MINE', 'MSW', 'NREN',
       'OZON', 'PEST', 'REN', 'SMOG', 'VADD', 'WATR'].sum()
df_new
```

```python
df_new.to_csv("data_US.csv", index=False)
```

Now an issue with above dataset is that if we look at a given county, it does not cover all the industries. 
So what we will do below is that we will fill null values for industries in the county that are not covered in above dataset.

# Second table with dummy values

```python
df1 = df[['fips', 'name', 'year']].drop_duplicates().sort_values(by=['fips']).reset_index(drop=True)
df1['code'] = [codes.tolist()] * len(df1)
df1
```

```python
df1 = df1.explode('code')
df1
```

So now the above DataFrame has 219800 rows, which is 3140 counties * 70 industry codes!

```python
df1.to_csv("unique.csv", index=False)
```

```python
# Filling with null values
vals = ['payann', 'estab', 'emp','ACID', 'ENRG', 'ETOX', 'EUTR', 'FOOD', 'GCC', 'HAPS', 'HAZW', 'HC', 
    'HNC', 'HRSP', 'HTOX', 'JOBS', 'LAND', 'METL', 'MINE', 'MSW', 'NREN',
       'OZON', 'PEST', 'REN', 'SMOG', 'VADD', 'WATR']
df1[vals] = pd.DataFrame([np.zeros(len(vals))], index=df1.index)
df1
```

Now, we concatenate the two tables.

```python
result = pd.concat([df_new, df1], sort=False)
result
```

```python
df_final = result.groupby(['fips', 'name', 'year', 'code'], as_index=False)['payann', 'estab', 'emp',
    'ACID', 'ENRG', 'ETOX', 'EUTR', 'FOOD', 'GCC', 'HAPS', 'HAZW', 'HC', 
    'HNC', 'HRSP', 'HTOX', 'JOBS', 'LAND', 'METL', 'MINE', 'MSW', 'NREN',
       'OZON', 'PEST', 'REN', 'SMOG', 'VADD', 'WATR'].sum()
df_final.head(62)
```

```python
# Keeping the social factors here, because we will ignore it for future pivot operation ..
social_factors = df_final.iloc[:,0:7]
social_factors
```

# Pivoting the table

```python
table = pd.pivot_table(df, values=['ACID', 'ENRG', 'ETOX', 'EUTR', 'FOOD', 'GCC', 'HAPS', 'HAZW', 'HC', 
    'HNC', 'HRSP', 'HTOX', 'JOBS', 'LAND', 'METL', 'MINE', 'MSW', 'NREN',
       'OZON', 'PEST', 'REN', 'SMOG', 'VADD', 'WATR'], index=['fips', 'name', 'year'],
                    columns=['code'], aggfunc=np.sum).fillna(0)
```

```python
table
```

```python
flattened = pd.DataFrame(table.to_records())
flattened
```

```python
# Rewriting column names
r = re.compile('(.*, .*)')
flattened.columns = [fi[-5:-2] + '_' + fi[2:fi.index(", ") - 1] if r.match(fi) is not None else fi 
 for fi in flattened.columns]
flattened
```

```python
#Rearranging the columns so that the input vector is arranged well

flattened = flattened.reindex(sorted(flattened.columns), axis=1)
flattened
```

Hence, each row represents data from a given county. If we look at a row, it coveres data columns in form of [XXX_YYYY] where XXX is the industry code, and YYYY is the USEEIO environmental feature.

```python
flattened.to_csv("modifiedData_US.csv", index=False)
```
