```python
import numpy as np
import pandas as pd
```

# Importing dataset and processing it

```python
df = pd.read_csv("cancer_industry.csv")
```

```python
df.head()
```

```python
new_df = pd.DataFrame(df['fips'].unique(), columns = ['fips'])
```

```python
new_df.head()
```

```python
# creating cancer columns:
for cancer in df['cancer'].unique():
    new_df[cancer] = 0
```

```python
new_df.head()
```

```python
new_df.set_index("fips", inplace=True)
```

```python
new_df.head()
```

```python
for index, row in df.iterrows():
    new_df.loc[row['fips'], row['cancer']] = row['incidence rate_per_100000']
```

```python
new_df.head()
```

```python
final_df = new_df.set_index(np.arange(new_df.shape[0]))
```

```python
new_df = new_df.reset_index()
```

```python
new_df.head()
```

```python
new_df.to_csv("rate_cancer_per_fips.csv", index=False)
```
